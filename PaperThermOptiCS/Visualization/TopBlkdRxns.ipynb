```python
# library to import .mat file
import scipy.io as sio
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
glc_rxns = ['GLCP','GLCP2','GLBRAN2','GLCS1','GLGC']
data = sio.loadmat('Results_ToFindTheRxnTBlkdInAllMeM_except_TOCS.mat')
rxns = data['ordered_rxns']
rxns = [rxn[0][0] for rxn in rxns]
size = data['ordered_size']
size = [s[0] for s in size]
# plotting a bar plot for the top 20 rxns
plt.figure(figsize=(15,10))
sns.barplot(x=rxns[:20], y=size[:20],color='darkorange')
# bold x-axis labels
plt.xticks(fontsize=20, fontweight='bold',rotation=90)
# bold y-axis labels
plt.yticks(fontsize=20, fontweight='bold')
plt.ylabel(f'Number of models the reaction is \n thermodynamically blocked',fontdict={'fontsize': 25, 'fontweight': 'bold'})
plt.xlabel('Reaction names',fontdict={'fontsize': 25, 'fontweight': 'bold'})
# coloring the xticks
for i,t in enumerate(plt.gca().xaxis.get_ticklabels()):
    if rxns[i] in glc_rxns:
        t.set_color('blue')
# adding the text on the top of the bars
for i in range(20):
    plt.text(i, size[i]+0.5, size[i], ha='center', va='bottom', fontsize=15, fontweight='bold')
# settign the spines off
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.savefig('Top20RxnsTBlkdInAllMeM_except_TOCS.pdf',bbox_inches='tight')
```
