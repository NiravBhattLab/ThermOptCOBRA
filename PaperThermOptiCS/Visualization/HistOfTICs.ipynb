```python
# library to import .mat file
import scipy.io as sio
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
TICDir = 'AGORA_TIC_Results'
# listing all the files in the directory
files = os.listdir(TICDir)
TIC_size=[]
for file in files:
    data = sio.loadmat(TICDir + '/' + file)
    TIC_size = TIC_size+[len(i[0]) for i in data['TICs']]
```

```python
# plotting the histogram with the pre-defined bins
b = 25
fig,ax = plt.subplots(1,1,figsize=(10,6))
bins = np.arange(0, b+1, 1)
sns.histplot(TIC_size, bins=bins, color='b', alpha=0.5, ax=ax)
ax.set_xticks(range(0,b+1,1))
plt.setp(ax.get_xticklabels(), fontweight='bold',fontsize=10)
plt.setp(ax.get_yticklabels(), fontweight='bold',fontsize=10)
ax.set_xlabel('Number of reaction in TICs',fontweight='bold',fontsize=14)
ax.set_ylabel('Count',fontweight='bold',fontsize=14)
ax.set_title(f'Histogram of TIC\'s size in \n models from AGORA database',fontweight='bold',fontsize=25)
ax.set_xlim([0,b])
# setting the spines off
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig('TIC_size_histogram_agora.pdf',bbox_inches='tight')

```
